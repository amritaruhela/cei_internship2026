```python
import pandas as pd
import sqlite3
df=pd.read_csv("/content/Sample - Superstore.csv",encoding="latin1")
print("Dataset Shape:",df.shape)
df.head()
```

```python
conn = sqlite3.connect("superstore.db")
df.to_sql(
    "superstore",
    conn,
    if_exists="replace",
    index=False
)
print("Table Created Successfully")
```

```python
query = """
PRAGMA table_info(superstore);
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM superstore
LIMIT 10;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS total_rows
FROM superstore;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM superstore
WHERE Region='South';
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM superstore
WHERE Category='Office Supplies';
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM superstore
WHERE Sales > 2367;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
Region,
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Region
ORDER BY Total_Sales DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
Category,
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Category
ORDER BY Total_Sales DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
Category,
AVG(Sales) AS Average_Sales,
SUM(Quantity) AS Total_Quantity
FROM superstore
GROUP BY Category;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
[Product Name],
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY [Product Name]
ORDER BY Total_Sales DESC
LIMIT 15;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
[Customer Name],
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY [Customer Name]
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
substr([Order Date],7,4) || '-' ||
substr([Order Date],1,2) AS Month,

SUM(Sales) AS Revenue
FROM superstore
GROUP BY Month
ORDER BY Month;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
[Order ID],
COUNT(*) AS duplicate_count

FROM superstore

GROUP BY [Order ID]

HAVING COUNT(*) > 1;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS Missing_Customers
FROM superstore
WHERE [Customer Name] IS NULL;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS Missing_Products
FROM superstore
WHERE [Product Name] IS NULL;
"""

pd.read_sql(query, conn)
```
