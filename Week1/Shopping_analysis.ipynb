```python
numerical_cols=csv_1.select_dtypes(include=['int64','float64']).columns
for col in numerical_cols:
  csv_1[col]=csv_1[col].fillna(csv_1[col].mean())
```

```python
categorical_cols=csv_1.select_dtypes(include=['object']).columns
for col in categorical_cols:
  csv_1[col]=csv_1[col].fillna(csv_1[col].mode()[0])
```

```python
selected_data=csv_1[['title','discount','final_price']]
selected_data.head()
```

```python
print(csv_1['final_price'].dtype)
print(csv_1['final_price'].head())
```

```python
csv_1['final_price']=(
    csv_1['final_price']
    .astype(str)
    .str.replace('"','',regex=False)
    .str.replace('₹','',regex=False)
    .str.replace(',','',regex=False)
)
csv_1['final_price']=pd.to_numeric(csv_1['final_price'],errors='coerce')
print(csv_1['final_price'].head())
print(csv_1['final_price'].dtype)
```

```python
filtered_data=csv_1[csv_1['final_price']>1500]
filtered_data.head()
```

```python
print("Before Duplicate",csv_1.shape)
csv_1.drop_duplicates(inplace=True)
print("After Duplicate",csv_1.shape)
```

```python
csv_1['quantity']=1.5
csv_1['total_amount']=csv_1['final_price']*csv_1['quantity']
csv_1[['title','final_price','quantity','total_amount']].head()
```

```python
csv_1.to_csv('cleaned_dataset.csv',index=False)
print("Cleaned dataset saved successfully!")
```
